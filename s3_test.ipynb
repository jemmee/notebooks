```python
import boto3
from moto import mock_aws
import os

BUCKET_NAME = 'my-unique-demo-bucket'
FILE_NAME = 'demo_object.txt'

@mock_aws
def run_s3_test():
    conn = boto3.client('s3', region_name='us-east-1')
    conn.create_bucket(Bucket=BUCKET_NAME)
    print(f"1. Created bucket: {BUCKET_NAME}")
    
    conn.put_object(Bucket=BUCKET_NAME, Key=FILE_NAME, Body=open(FILE_NAME).read())
    print(f"2. Uploaded object: {FILE_NAME}")

    print(f"3. Listing objects in '{BUCKET_NAME}':")
    response = conn.list_objects_v2(Bucket=BUCKET_NAME)
    
    if 'Contents' in response:
        for obj in response['Contents']:
            print(f"   - Found: {obj['Key']} (Size: {obj['Size']} bytes)")
    else:
        print("   - Bucket is empty.")

    response = conn.get_object(Bucket=BUCKET_NAME, Key=FILE_NAME)
    downloaded_data = response['Body'].read().decode('utf-8')
    open(FILE_NAME, 'w').write(downloaded_data)
    print(f"4. Downloaded: '{FILE_NAME}'")

run_s3_test()
```
