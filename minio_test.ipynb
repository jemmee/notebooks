```python
# docker run \
#  -p 9000:9000 \
#  -p 9001:9001 \
#  --name minio \
#  -v ~/minio/data:/data \
#  -e "MINIO_ROOT_USER=minioadmin" \
#  -e "MINIO_ROOT_PASSWORD=minioadmin" \
#  quay.io/minio/minio server /data --console-address ":9001"

import io
from minio import Minio

# Initialize client
client = Minio(
    "localhost:9000",
    access_key="minioadmin",
    secret_key="minioadmin",
    secure=False # Required for local HTTP connections
)

# Create a bucket if it doesn't exist
bucket_name = "test-bucket"
if not client.bucket_exists(bucket_name):
    client.make_bucket(bucket_name)
    print(f"Created bucket: {bucket_name}")

# 1. Convert your bytes into a stream/file-like object
data_to_upload = b"Hello MinIO!"
data_stream = io.BytesIO(data_to_upload)

# 2. Upload using the stream
client.put_object(
    bucket_name, 
    "hello.txt", 
    data_stream, 
    length=len(data_to_upload),
    content_type="text/plain" # Good practice to specify this
)

print("File uploaded successfully!")
```
