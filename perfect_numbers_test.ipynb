```python
def is_perfect(n):
    if n <= 1:
        return False
    
    divisor_sum = 1  # 1 is always a proper divisor
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            divisor_sum += i
            if i != n // i and i != 1:  # Avoid adding sqrt twice, skip 1 again
                divisor_sum += n // i
    return divisor_sum == n
    
def find_perfect_numbers(limit):
    perfect_numbers = []
    for num in range(2, limit + 1):
        if is_perfect(num):
            perfect_numbers.append(num)
    return perfect_numbers

# Find all known small perfect numbers
print(find_perfect_numbers(10000))
# Output: [6, 28, 496, 8128]

#print(find_perfect_numbers(100000000))
```
