```python
# docker run -d \
#  --name trino \
#  -p 8080:8080 \
#  trinodb/trino

from trino.dbapi import connect

# 1. Setup the connection
# Note: 'user' is required by Trino but can be any string for this demo
conn = connect(
    host="localhost",
    port=8080,
    user="admin",
    catalog="tpch",
    schema="tiny",
)

# 2. Create a cursor to execute queries
cur = conn.cursor()

print("--- Querying Top 5 Nations by Region ---")
query = """
    SELECT n.name, r.name as region 
    FROM nation n 
    JOIN region r ON n.regionkey = r.regionkey 
    LIMIT 5
"""

try:
    # 3. Execute the query
    cur.execute(query)
    
    # 4. Fetch and print results
    rows = cur.fetchall()
    for row in rows:
        print(f"Nation: {row[0]:<15} | Region: {row[1]}")

except Exception as e:
    print(f"Error: {e}")

finally:
    # 5. Always close your connections
    cur.close()
    conn.close()
```
