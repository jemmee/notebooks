```python
from pysnmp.hlapi import *

def snmp_get_sys_descr(target_ip, community='public'):
    # Object Identity for System Description (sysDescr)
    oid = ObjectIdentity('1.3.6.1.2.1.1.1.0')

    # Generator for the SNMP GET request
    iterator = getCmd(
        SnmpEngine(),
        CommunityData(community, mpModel=0), # mpModel=0 is SNMPv1, 1 is v2c
        UdpTransportTarget((target_ip, 161)),
        ContextData(),
        ObjectType(oid)
    )

    errorIndication, errorStatus, errorIndex, varBinds = next(iterator)

    # Error handling for the SIEM log
    if errorIndication:
        print(f"Connection Error: {errorIndication}")
    elif errorStatus:
        print(f"SNMP Error: {errorStatus.prettyPrint()}")
    else:
        for varBind in varBinds:
            print(f"âœ… Success! Device Info: {' = '.join([x.prettyPrint() for x in varBind])}")

# Demo (using a common public tester IP or a local device)
snmp_get_sys_descr('localhost')
```
