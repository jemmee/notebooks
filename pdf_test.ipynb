```python
from fpdf import FPDF

# Create a new PDF object
pdf = FPDF()

# Add a page
pdf.add_page()

# Set the font (e.g., Arial, size 12)
pdf.set_font("Arial", size=12)

# Add text to the PDF
# cell(width, height, text, border, next_line, align)
pdf.cell(200, 10, txt="Hello, this is a PDF generated with Python!", ln=True, align='C')

# Add another line of text
pdf.cell(200, 10, txt="This is a second line of text.", ln=True, align='L')

# Save the PDF to a file
pdf.output("my_first_pdf.pdf")

print("PDF generated successfully!")
```

```python
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter

# Create a Canvas object
# 'hello.pdf' is the output filename
c = canvas.Canvas("hello_reportlab.pdf", pagesize=letter)

# Set the font and size
c.setFont("Helvetica", 14)

# Draw a string at a specific position (x, y coordinates from bottom-left)
c.drawString(100, 750, "Hello, ReportLab!")
c.drawString(100, 730, "This is a more precise way to add content.")

# Draw a geometric shape (a red rectangle)
c.setFillColorRGB(1, 0, 0) # Red color
c.rect(100, 700, 150, 20, fill=1) # (x, y, width, height, fill)

# Save the PDF file
c.showPage() # Ends the current page
c.save()

print("PDF generated successfully!")
```
