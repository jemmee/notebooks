```python
# docker run --name my-db -e POSTGRES_PASSWORD=mysecret -p5432:5432 -d postgres

import psycopg2

# 1. Connection Details
# If using Docker, ensure you have '-p 5432:5432' in your run command
connection_params = {
    "host": "localhost",
    "database": "postgres",
    "user": "postgres",
    "password": "mysecret", # The one you set in Docker
    "port": "5432"
}

try:
    # 2. Establish Connection
    print("Connecting to PostgreSQL...")
    conn = psycopg2.connect(**connection_params)
    cursor = conn.cursor()

    # 3. Create a Test Table
    print("Creating test table...")
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS test_table (
            id SERIAL PRIMARY KEY,
            message TEXT,
            created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
        );
    """)

    # 4. Insert Data
    print("Inserting test record...")
    cursor.execute("INSERT INTO test_table (message) VALUES (%s)", ("Hello from Python!",))
    
    # 5. Commit changes (Postgres won't save without this!)
    conn.commit()

    # 6. Read Data back
    print("Verifying data...")
    cursor.execute("SELECT * FROM test_table ORDER BY created_at DESC LIMIT 1;")
    record = cursor.fetchone()
    print(f"Success! Database says: {record}")

except Exception as e:
    print(f"Error: {e}")

finally:
    # 7. Clean up
    if 'cursor' in locals(): cursor.close()
    if 'conn' in locals(): conn.close()
    print("Connection closed.")
```
