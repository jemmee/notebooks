```python
import qrcode

# The data you want to encode
data = input("Paste your data: ")

qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)

qr.add_data(data)
qr.make(fit=True)

# 3. Create and save the image
img = qr.make_image(fill_color="black", back_color="white")
filename = "my_qr_code.png"
img.save(filename)

print(f"\n✅ Success! Your QR code is saved as {filename}")
```
